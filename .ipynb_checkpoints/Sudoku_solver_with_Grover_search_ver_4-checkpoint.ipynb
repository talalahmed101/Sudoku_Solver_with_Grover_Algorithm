Determining Cell index and box index for a cell in Sudoku grid

```python
# Given the Cell of Sudoku grid with row 'r' and column 'c', identifying the associated box index
def box_row(r, n):
    return (r - 1) // n

def box_col(c, n):
    return (c - 1) // n

def box_index(r, c, n):
    return box_row(r, n) * n + box_col(c, n)

def cell_index(r, c, n):
    return (r - 1) * (n ** 2) + (c - 1)
```

Sudoku grid to cell value dictionary

```python
# Grid to dict {cell_index: int_value}
def grid_to_cellvalue_dict(grid, n):
    """
    grid: 2D list of size (n^2) x (n^2). Entries can be ints; empty can be 0 or None.
    Returns: dict {cell_index: int_value} with empty -> 0.
    """
    N = n * n
    if len(grid) != N or any(len(row) != N for row in grid):
        raise ValueError(f"grid must be {N}x{N} for n={n}")

    d = {}
    for r in range(1, N + 1):  # 1-based for cell_index
        for c in range(1, N + 1):
            v = grid[r - 1][c - 1]
            d[cell_index(r, c, n)] = 0 if (v is None or v == 0 or v == '0') else int(v)
    return d

# empty cell_index to positions from cellvalue_dict
def empty_index_to_position_map(cellvalue_dict):
    """
    cellvalue_dict: {cell_index: value}, where empty cells have value 0.
    Returns: {empty_cell_index: position}, where position runs 0..K-1
             in increasing order of empty_cell_index.
    """
    empty_indices = sorted([i for i, v in cellvalue_dict.items() if v == 0])
    return {idx: pos for pos, idx in enumerate(empty_indices)}
```

Collecting Row, Column and Box constraints

```python
def row_constraints_dicts(grid):
    """
    For each row that contains >=1 empty cell (value 0 or None),
    build a dict where:
      - empty cells (by cell_index) come first (sorted by cell_index), mapped to 0
      - then filled cells (sorted by cell_index), mapped to their integer values
    Returns: list of row dictionaries (top row to bottom row).
    """
    N = len(grid)
    if any(len(row) != N for row in grid):
        raise ValueError("Grid must be square.")
    n = int(math.isqrt(N))
    if n * n != N:
        raise ValueError("Grid size must be n^2 x n^2 (e.g., 4x4, 9x9, 16x16).")

    out = []
    for r in range(1, N + 1):
        empties, filled = [], []
        for c in range(1, N + 1):
            idx = cell_index(r, c, n)
            v = grid[r - 1][c - 1]
            if v is None or v == 0:
                empties.append(idx)
            else:
                filled.append((idx, int(v)))

        if not empties:
            continue

        empties.sort()
        filled.sort(key=lambda t: t[0])

        d = {}
        for idx in empties:
            d[idx] = 0
        for idx, v in filled:
            d[idx] = v

        out.append(d)

    return out
```

```python
def col_constraints_dicts(grid):
    """
    For each column that contains >=1 empty cell (value 0 or None),
    build a dict where:
      - empty cells (by cell_index) come first (sorted by cell_index), mapped to 0
      - then filled cells (sorted by cell_index), mapped to their integer values
    Returns: list of column dictionaries (left column to right column).
    """
    N = len(grid)
    if any(len(row) != N for row in grid):
        raise ValueError("Grid must be square.")
    n = int(math.isqrt(N))
    if n * n != N:
        raise ValueError("Grid size must be n^2 x n^2 (e.g., 4x4, 9x9, 16x16).")

    out = []
    for c in range(1, N + 1):
        empties, filled = [], []
        for r in range(1, N + 1):
            idx = cell_index(r, c, n)
            v = grid[r - 1][c - 1]
            if v is None or v == 0:
                empties.append(idx)
            else:
                filled.append((idx, int(v)))

        if not empties:
            continue

        empties.sort()
        filled.sort(key=lambda t: t[0])

        d = {}
        for idx in empties:
            d[idx] = 0
        for idx, v in filled:
            d[idx] = v

        out.append(d)

    return out
```

```python
def box_constraints_dicts(grid):
    """
    For each box that contains >=1 empty cell (0 or None),
    build a dict where:
      - empty cells (by cell_index) come first (sorted), mapped to 0
      - then filled cells (sorted), mapped to their integer values
    Returns: list of box dictionaries in increasing box_index order,
             skipping boxes with no empty cells.
    """
    N = len(grid)
    if any(len(row) != N for row in grid):
        raise ValueError("Grid must be square.")
    n = int(math.isqrt(N))
    if n * n != N:
        raise ValueError("Grid size must be n^2 x n^2 (e.g., 4x4, 9x9, 16x16).")

    out = []
    num_boxes = N  # (n^2) boxes total

    for b in range(num_boxes):  # 0..N-1
        br = b // n  # box row (0..n-1)
        bc = b % n   # box col (0..n-1)

        empties, filled = [], []

        r_start = br * n + 1
        c_start = bc * n + 1

        for r in range(r_start, r_start + n):
            for c in range(c_start, c_start + n):
                idx = cell_index(r, c, n)
                v = grid[r - 1][c - 1]
                if v is None or v == 0:
                    empties.append(idx)
                else:
                    filled.append((idx, int(v)))

        if not empties:
            continue  # only include boxes with empties

        empties.sort()
        filled.sort(key=lambda t: t[0])

        d = {}
        for idx in empties:
            d[idx] = 0
        for idx, v in filled:
            d[idx] = v

        out.append(d)

    return out
```

Quantum Register and Circuit for Sudoku grid

```python
def sudoku_quantum_register_circuit(n,
                                 empty_index_to_position_map,
                                 total_constraints):
    """
    Creates a Qiskit circuit with registers:
      - x:    N = K*d qubits for K empty cells, grouped by d per empty cell
      - b:    n^2 bucket qubits
      - c:   c_tot qubits for total constraints
      - anc:  1 global ancilla
    """
    Nside = n * n
    d = max(1, math.ceil(math.log2(Nside)))  # qubits per cell value
    K = len(empty_index_to_position_map)
    N = K * d

    c_tot = len(total_constraints)

    tot_qubits = N + Nside + c_tot + 1

     # Registers (order here is the qubit order in the circuit)
    x  = QuantumRegister(N,      "x")    # variable register: K groups of d
    b  = QuantumRegister(Nside,  "b")    # bucket register: n^2
    c = QuantumRegister(c_tot,    "c")   # total constraints   
    anc = QuantumRegister(1,     "anc")  # global ancilla

    qc = QuantumCircuit(x, b, c, anc)

    # Map each empty cell index -> its d qubits in x
    # (position order is given by empty_index_to_position_map values 0..K-1)
    
    empty_cell_to_qubits = {}
    for cell_idx, pos in sorted(empty_index_to_position_map.items(), key=lambda kv: kv[1]):
        start = pos * d
        empty_cell_to_qubits[cell_idx] = [x[start + j] for j in range(d)]

    return qc, empty_cell_to_qubits, x, b, c, anc, tot_qubits
```

```python
def sudoku_quantum_register_circuit_oracle(n,
                                 empty_index_to_position_map,
                                 total_constraints):
    """
    Creates a Qiskit circuit with registers:
      - x:    N = K*d qubits for K empty cells, grouped by d per empty cell
      - b:    n^2 bucket qubits
      - c:   c_tot qubits for total constraints
      - anc:  1 global ancilla
    """
    Nside = n * n
    d = max(1, math.ceil(math.log2(Nside)))  # qubits per cell value
    K = len(empty_index_to_position_map)
    N = K * d

    c_tot = len(total_constraints)

    tot_qubits = N + Nside + c_tot + 1

     # Registers (order here is the qubit order in the circuit)
    x  = QuantumRegister(N,      "x")    # variable register: K groups of d
    b  = QuantumRegister(Nside,  "b")    # bucket register: n^2
    c = QuantumRegister(c_tot,    "c")   # total constraints   
    anc = QuantumRegister(1,     "anc")  # global ancilla

    qc = QuantumCircuit(x, b, c, anc, name = "Oracle")

    return qc
```

```python
def bits_lsb_first(a, d):
    # returns [a0, a1, ..., a_{n-1}]
    return [(a >> j) & 1 for j in range(d)]
```

```python
def d_qubits(n):
    # n^2 symbols in each row/col/box
    return max(1, math.ceil(math.log2(n * n)))
```

```python
def empty_keys_from_constraint(constraint_dict):
    """
    constraint_dict like {13:0, 14:0, 12:2, 15:1}
    Returns the empty cell_indices in increasing order.
    """
    return sorted([k for k, v in constraint_dict.items() if v == 0])

def nonzero_items_from_constraint(constraint_dict):
    """
    Returns:
      - nz_keys:   list of cell_indices with non-zero values (sorted)
      - nz_values: list of the corresponding non-zero values in the same order
      - nz_dict:   dict of only the non-zero entries
    """
    nz_items = sorted([(k, v) for k, v in constraint_dict.items() if v != 0], key=lambda t: t[0])
    nz_keys = [k for k, _ in nz_items]
    nz_values = [v for _, v in nz_items]
    nz_dict = {k: v for k, v in nz_items}
    return nz_keys, nz_values, nz_dict

def nonzero_values_from_constraint(constraint_dict):
    """
    Returns:
      - nz_keys:   list of cell_indices with non-zero values (sorted)
      - nz_values: list of the corresponding non-zero values in the same order
      - nz_dict:   dict of only the non-zero entries
    """
    nz_items = sorted([(k, v) for k, v in constraint_dict.items() if v != 0], key=lambda t: t[0])
    nz_values = [v for _, v in nz_items]
    return nz_values
```

Bit-mask checking and constructing Oracle

```python
def bitmask_check_for_single_constraint(qc0, x_reg, b_reg, empty_cell_to_qubits, constraint, n, d):
    qc = qc0.copy()
    empty_cell_keys = empty_keys_from_constraint(constraint)
    non_zero_cell_values = nonzero_values_from_constraint(constraint)
    bucket = list(range(n**2))
   
    for k in range(len(bucket)):
        cell_value = bucket[k]
        bit_list = bits_lsb_first(cell_value, d)
        b_reg_cell = b_reg[cell_value]
        for i in empty_cell_keys:
            ctrl_qubits = empty_cell_to_qubits[i] 
            for j in range(len(ctrl_qubits)):
                if bit_list[j] == 0:
                    qc.x(ctrl_qubits[j])
            qc.mcx(ctrl_qubits, b_reg_cell)
            for j in range(len(ctrl_qubits)):
                if bit_list[j] == 0:
                    qc.x(ctrl_qubits[j])
        #qc.barrier()
    for p in range(len(non_zero_cell_values)):
        nz_value = non_zero_cell_values[p]-1
        qc.x(b_reg[nz_value])
    #qc. barrier()
    return qc
```

```python
def bitmask_check_for_single_constraint_extended(qc0, x_reg, b_reg, c_reg, empty_cell_to_qubits, constraint, constr_no, n, d):
    qc = qc0.copy()
    logic_circuit = QuantumCircuit(x_reg, b_reg)
    logic_circuit = bitmask_check_for_single_constraint(logic_circuit, x_reg, b_reg, empty_cell_to_qubits, constraint, n, d)
    qc.compose(logic_circuit, inplace=True)
    qc.mcx(b_reg, c_reg[constr_no])
    qc.compose(logic_circuit.inverse(), inplace=True)
    #qc2.barrier()
    return qc
```

```python
def bitmask_check_for_multi_constraints(qc, x_reg, b_reg, c_reg, empty_cell_to_qubits, multi_constraints, n, d):
  #  qc = qc0.copy()
    cq = len(multi_constraints)
    for i in range(cq):
        qc = bitmask_check_for_single_constraint_extended(qc, x_reg, b_reg, c_reg, empty_cell_to_qubits, multi_constraints[i], i, n, d)
        #qc.barrier()
    return qc
        
```

```python
def bitmask_check_for_multi_constraints_reversed(qc0, x_reg, b_reg, c_reg, empty_cell_to_qubits, multi_constraints, n, d):
    qc = qc0.copy()
    qc1 = bitmask_check_for_multi_constraints(qc, x_reg, b_reg, c_reg, empty_cell_to_qubits, multi_constraints, n, d)
    return qc1.inverse() 
        
```

```python
def bitmask_check_for_all_constraints_oracle(qc0, x_reg, b_reg, c_reg, anc_reg, empty_cell_to_qubits, total_constraints, n, d):
    qc_01 = qc0.copy()
    qc_02 = qc0.copy()
    qc_1 = bitmask_check_for_multi_constraints(qc_01, x_reg, b_reg, c_reg, empty_cell_to_qubits, total_constraints, n, d)
    qc_2 = bitmask_check_for_multi_constraints_reversed(qc_02, x_reg, b_reg, c_reg, empty_cell_to_qubits, total_constraints, n, d)
    qc_1.mcx(c_reg, anc_reg[0])
    qc3 = qc_1.compose(qc_2)
    return qc3
```

```python
def sudoku_qubit_count(empty_cell, d):
    return empty_cell*d

def sudoku_search_dim(empty_cell, d):
    return 2**(sudoku_qubit_count(empty_cell,d))

def Grover_iteration(N):
    return round(np.sqrt(N)*(pi/4)-1/2)
```

```python
def Grover_Diffuser_circuit(x):
    dq = len(x)
    qr_d = QuantumRegister(dq, 'x')
    qc_d0 = QuantumCircuit(qr_d, name = "Diffuser")
    qc_d = qc_d0.copy()
    for i in range(dq):
        qc_d.h(x[i])
        qc_d.x(x[i])
    qc_d.h(x[dq-1])
    qc_d.mcx([x[k] for k in range(dq-1)], x[dq-1])
    qc_d.h(x[dq-1])
    for i in range(dq):
        qc_d.x(x[i])
        qc_d.h(x[i])
    return qc_d
```

```python
def Grover_Full_Diffusion(qc, x, anc, depth, oracle_gate, Diffuse_gate):
    qc_g = qc.copy()
    qc_g.x(anc[0])
    qc_g.h(anc[0])
    for i in range(len(x)):
        qc_g.h(x[i])
    for j in range(depth):
        qc_g.compose(oracle_gate, inplace = True)
        qc_g.compose(Diffuse_gate, inplace = True)
    return qc_g
```

```python
def total_empty_cells(cell_dict):
    k = 0
    for key, value in cell_dict.items():
        if value == 0:
            k = k +1
    return k

def expected_values(dict_1, dict_2):
    result = {key: dict_1[key] for key, value in dict_2.items() if value == 0}
    return result

def expected_values_to_bitstring(filtered_dict, d):
    bitstring = []
    for val in filtered_dict.values():
        bits = bits_lsb_first(val - 1, d)
        bitstring.extend(bits)
        bitstring1 = bitstring[::-1]
        bitstring2 = [str(k) for k in bitstring1]
    return "".join(bitstring2)
```

Statevector-based verification

```python
def qubit_indices(qc, reg):
    return [qc.find_bit(q).index for q in reg]


def minus_state():
    return Statevector(np.array([1, -1], dtype=complex) / np.sqrt(2))


def expected_x_state(empty_map, expected_cell_values, d):
    """
    Build the expected basis state on the x-register only.
    """
    K = len(empty_map)
    qc_id = QuantumCircuit(K * d)

    for cell_idx, pos in sorted(empty_map.items(), key=lambda kv: kv[1]):
        encoded_val = expected_cell_values[cell_idx] - 1
        bits = bits_lsb_first(encoded_val, d)
        for j, bit in enumerate(bits):
            if bit == 1:
                qc_id.x(pos * d + j)

    return Statevector.from_instruction(qc_id)


def print_top_probabilities(prob_dict, top_k=8):
    items = sorted(prob_dict.items(), key=lambda kv: kv[1], reverse=True)
    for bitstr, prob in items[:top_k]:
        print(f"{bitstr}: {prob:.12f}")
```

Verification

```python
# Exact statevector-based verification for the 4x4 example

sv = Statevector.from_instruction(qc_g)

x_idx = qubit_indices(qc_g, x)
b_idx = qubit_indices(qc_g, b)
c_idx = qubit_indices(qc_g, c)
anc_idx = qubit_indices(qc_g, anc)

# Exact marginal probabilities on the x-register
x_probs = sv.probabilities_dict(qargs=x_idx)

# Reduced states
rho_x = partial_trace(sv, b_idx + c_idx + anc_idx)
rho_bc = partial_trace(sv, x_idx + anc_idx)
rho_anc = partial_trace(sv, x_idx + b_idx + c_idx)

# Expected states
ideal_x = expected_x_state(empty_map, expected_cell_values, d)
zero_bc = DensityMatrix.from_label("0" * (len(b) + len(c)))
minus_anc = minus_state()

# Fidelities
fid_x = state_fidelity(rho_x, ideal_x)
fid_bc = state_fidelity(rho_bc, zero_bc)
fid_anc = state_fidelity(rho_anc, minus_anc)


print("Top exact x-register probabilities:")
print_top_probabilities(x_probs, top_k=8)

largest_key = max(x_probs, key=x_probs.get)

print()
print(f"Most likely x-register state                  : {largest_key}")
print(f"Expected x-register state                     : {expected_bitstring}")
print(f"Exact probability of expected bitstring       : {x_probs.get(expected_bitstring, 0.0):.12f}")
print(f"x-register fidelity with target state         : {fid_x:.12f}")
print(f"Oracle Workspace reset fidelity (b + c)       : {fid_bc:.12f}")
print(f"Ancilla fidelity with |->                     : {fid_anc:.12f}")
```

```python
qc_g3 = Grover_Full_Diffusion(qc3, x3, anc3, Gr_it3, oracle_gate3, Diffuse_gate3)
```

```python
# Exact statevector-based verification for the 9x9 example

sv3 = Statevector.from_instruction(qc_g3)

x_idx3 = qubit_indices(qc_g3, x3)
b_idx3 = qubit_indices(qc_g3, b3)
c_idx3 = qubit_indices(qc_g3, c3)
anc_idx3 = qubit_indices(qc_g3, anc3)

# Exact marginal probabilities on the x-register
x_probs3 = sv3.probabilities_dict(qargs=x_idx3)

# Reduced states
rho_x3 = partial_trace(sv3, b_idx3 + c_idx3 + anc_idx3)
rho_bc3 = partial_trace(sv3, x_idx3 + anc_idx3)
rho_anc3 = partial_trace(sv3, x_idx3 + b_idx3 + c_idx3)

# Expected states
ideal_x3 = expected_x_state(empty_map3, expected_cell_values3, d3)
zero_bc3 = DensityMatrix.from_label("0" * (len(b3) + len(c3)))
minus_anc3 = minus_state()

# Fidelities
fid_x3 = state_fidelity(rho_x3, ideal_x3)
fid_bc3 = state_fidelity(rho_bc3, zero_bc3)
fid_anc3 = state_fidelity(rho_anc3, minus_anc3)


print("Top exact x-register probabilities:")
print_top_probabilities(x_probs3, top_k=8)

largest_key3 = max(x_probs3, key=x_probs3.get)

print()
print(f"Most likely x-register state                  : {largest_key3}")
print(f"Expected x-register state                     : {expected_bitstring3}")
print(f"Exact probability of expected bitstring       : {x_probs.get(expected_bitstring3, 0.0):.12f}")
print(f"x-register fidelity with target state         : {fid_x3:.12f}")
print(f"Oracle Workspace reset fidelity (b + c)       : {fid_bc3:.12f}")
print(f"Ancilla fidelity with |->                     : {fid_anc3:.12f}")
```
